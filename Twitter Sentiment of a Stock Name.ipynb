```python
# Take a ticker for any stock (example: tsla) and search all the tweets about it
search_results = twitter.search.tweets(q = 'tsla', count = 100)
statuses = search_results['statuses']
df = json_normalize(search_results, 'statuses')
status_texts = df["text"]
status_texts
```

```python
# Calculate polarity of tweets
polarity = []
for x in status_texts:
    polarity.append(TextBlob(x).sentiment.polarity)
polarity
len(polarity)
```

```python
polarity
```

```python
# Get Screen Names
screen_names = [status['user']['screen_name']
               for status in statuses]
screen_names
```

```python
# Identify all the screen names that are positive (.7 or greaters) and negative (-.7 and less)
```

```python
# Polarity 0.7 or greater
screen_names_high = [status['user']['screen_name']
               for status in statuses
                    if TextBlob(status['text']).sentiment.polarity >= 0.7]
screen_names_high
```

```python
# Polarity -0.7 or lower
screen_names_low = [status['user']['screen_name']
               for status in statuses
                    if TextBlob(status['text']).sentiment.polarity >= -0.7]                    
screen_names_low
```

```python
# Create a new data frame of Ploarity, Screen Names and Tweets
df_final = pd.DataFrame({"Polarity" : polarity, "Screen Name" : screen_names, "Tweets" : status_texts})
df_final
```

```python
# Identify highest and lowest user based on polarity
user_highest = df_final.loc[df_final["Polarity"].idxmax()]["Screen Name"]
user_highest
```

```python
user_lowest = df_final.loc[df_final["Polarity"].idxmin()]["Screen Name"]
user_lowest
```
