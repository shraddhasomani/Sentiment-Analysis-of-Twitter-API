```python
# Search for a term
q = '#SuperBowl'
count = 250
```

```python
search_results = twit_api.search.tweets(q = q, count = count)
statuses = search_results['statuses']
statuses[0]['text']
```

```python
# Get screen_names
status_texts = [status['text'] for status in statuses]
screen_names = [user_mention['screen_name']
                for status in statuses
                    for user_mention in status['entities']['user_mentions']]
screen_names
```

```python
# Get hashtages
hashtags = [hashtag['text']
            for status in statuses
                for hashtag in status['entities']['hashtags']]
hashtags
```

```python
# Split the words
words = [w
        for t in status_texts
            for w in t.split()]
words
```

```python
from collections import Counter
for item in [words, screen_names, hashtags]:
    c = Counter(item)
    print(c.most_common()[:10])
    print()
```

```python
pip.main(["install", "prettytable"])
from prettytable import PrettyTable

for label, data in (('Word', words), ('Screen Name', screen_names), ('Hashtag', hashtags)):
    pt = PrettyTable(field_names = [label, 'Count'])
    c = Counter(data)
    [pt.add_row(kv) for kv in c.most_common()[:10]]
    pt.align[label], pt.align['Count'] = 'l', 'r'
    print(pt)
```

```python
# Retweets Count
retweets = [(status['retweet_count'], status['retweeted_status']['user']['screen_name'], status['text'])
            for status in statuses
                if status['retweet_count'] > 11]
retweets
```

```python
pt = PrettyTable(field_names = ['Count', 'Screen Name', 'Text'])
[pt.add_row(row) for row in sorted(retweets, reverse = True)[:5]]
pt.max_width['Text'] = 50
pt.align = 'l'
print(pt)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib.pyplot as plt
word_counts = sorted(Counter(words).values(), reverse = True)
plt.loglog(word_counts)
plt.ylabel("Freq")
plt.xlabel("Word Rank")
```

```python
for label, data in (('Words', words),('Screen Names', screen_names),('Hashtags', hashtags)):
        c = Counter(data)
        plt.hist(list(c.values()))
        plt.title(label)
        plt.ylabel("No of items in bin")
        plt.xlabel("Bins (no of times an item appeared)")
        plt.figure()
```
